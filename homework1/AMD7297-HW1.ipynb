```python
# Monte Carlo Pi approximation with error analysis
# Citation:
# Monte Carlo method for estimating Pi via random sampling.
# Reference: https://en.wikipedia.org/wiki/Monte_Carlo_method
# Reference: chatgpt

import random
import math
import numpy as np
import matplotlib.pyplot as plt

# Monte Carlo Pi approximation function
def monte_carlo_pi(n_samples):
    """
    Approximates the value of Pi using the Monte Carlo method.

    Parameters
    n_samples : int
        Number of random points to generate.

    Returns
    float
        Estimated value of Pi.
    """
    inside_circle = 0

    for _ in range(n_samples):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <= 1.0:
            inside_circle += 1

    return 4.0 * inside_circle / n_samples

# Main computation, iterate sample sizes and compute errors
sample_sizes = range(10, 10001, 20)
errors = []

# True value of Pi from Python, rounded to 15 decimal places
true_pi = round(math.pi, 15)

for n in sample_sizes:
    pi_estimate = monte_carlo_pi(n)
    error = abs(pi_estimate - true_pi)
    errors.append(error)

# Plot routine
plt.figure()
plt.plot(sample_sizes, errors)
plt.xlabel("Number of Samples")
plt.ylabel("Absolute Error")
plt.title("Monte Carlo Pi Approximation Error vs. Sample Size")
plt.show()
```
