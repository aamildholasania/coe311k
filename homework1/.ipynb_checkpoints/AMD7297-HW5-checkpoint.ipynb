```python
#Reference: chatgpt, I asked to confirm least square approach, sorting function, how to print residuals

import numpy as np
import matplotlib.pyplot as plt

#Custom Least Squares Function
def least_squares_fit(x_vals, y_vals):
    n_points = len(x_vals)
    sum_x = np.sum(x_vals)
    sum_y = np.sum(y_vals)
    sum_x_squared = np.sum(x_vals ** 2)
    sum_xy = np.sum(x_vals * y_vals)
    slope = (n_points * sum_xy - sum_x * sum_y) / (n_points * sum_x_squared - sum_x ** 2)
    intercept = (sum_y - slope * sum_x) / n_points
    return slope, intercept


# testing with rand vectors
x_random = np.random.rand(100)
y_random = np.random.rand(100)

# sort x values
order = np.argsort(x_random)
x_random = x_random[order]
y_random = y_random[order]

print("Random X values:", x_random)
print("Random Y values:", y_random)

rand_slope, rand_intercept = least_squares_fit(x_random, y_random)

print("Results from custom least squares on random data")
print("Slope:", rand_slope)
print("Intercept:", rand_intercept)


#Linear Data
x = np.arange(0, 1.01, 0.01)
true_slope = np.random.rand()
Y = m * X1 + b + np.random.rand(X1.size)

print("Original Linear Model")
print("Slope:", true_slope)
print("Intercept:", true_intercept)


#custom least squares
calc_slope, calc_intercept = least_squares_fit(x, y)

print("Custom Least Squares Result")
print("Calculated slope:", calc_slope)
print("Calculated intercept:", calc_intercept)


#compare with NumPy
numpy_slope, numpy_intercept = np.polyfit(x, y, 1)

print("NumPy Regression Result")
print("Slope:", numpy_slope)
print("Intercept:", numpy_intercept)


#print residuals
residuals = y - (calc_slope * x + calc_intercept)
largest_error_index = np.argmax(np.abs(residuals))
smallest_error_index = np.argmin(np.abs(residuals))
print("Largest Residual:", residuals[largest_error_index], "at x =", x[largest_error_index])
print("Smallest Residual:", residuals[smallest_error_index], "at x =", x[smallest_error_index])

print("All Residuals:")
for r in residuals:
    print(r)


#Plot
plt.figure()

plt.scatter(x, y, label="Data Points")
plt.plot(x, calc_slope * x + calc_intercept, label="Least Squares Fit")

plt.title("Linear Regression using Custom Least Squares")
plt.xlabel("X")
plt.ylabel("Y")

plt.legend()
plt.show()
```
