```python
import numpy as np
import matplotlib.pyplot as plt

#function
f = lambda x: np.sin(x)

#interval and points
a = 0
b = 2 * np.pi
N = 50

# Create grid
x = np.linspace(a, b, N)

# Step size
h = (b - a) / (N - 1)

#central difference
df_central = (f(x + h) - f(x - h)) / (2 * h)

#forward difference
df_forward = (f(x + h) - f(x)) / h

#backward difference
df_backward = (f(x) - f(x - h)) / h

# True derivative
df_exact = np.cos(x)

plt.figure(figsize=(10, 6))

plt.plot(x, df_exact, label='Exact (cos(x))', linewidth=3)
plt.plot(x, df_forward, '--', label='Forward Difference')
plt.plot(x, df_backward, ':', label='Backward Difference')
plt.plot(x, df_central, '-.', label='Central Difference')

plt.title("Derivative Approximation Comparison")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.legend()
plt.grid(True)

plt.show()
```
