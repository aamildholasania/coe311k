```python
import random
import numpy as np
import matplotlib.pyplot as plt

def approximatePiPlotted(numShots):
    counter = 0
    approximateError = np.zeros(100)
    relativeError = np.zeros(100)
    for j in range(0, 100):
        for i in range(0, numShots):
            x = random.uniform(-1, 1)
            y = random.uniform(-1, 1)
            if (x*x + y*y < 1):
                counter += 1
        piApproximate = 4.0 * counter / numShots
        approximateError[j] = abs(piApproximate - np.pi)
        relativeError[j] = abs((piApproximate - np.pi) / np . pi)
        counter = 0
    return (approximateError, relativeError)
x, y = approximatePiPlotted(1000)
t = np.arange(1, 101, 1)

plt.plot(t, x, label="approximate error")
plt.plot(t, y, label="relative error")
plt.legend()
plt.title("monte carlo error bounds")
plt.xlabel('num iterations')
plt.ylabel('error')
plt.show()
```
