```python
import numpy as np
from time import perf_counter

# data
A = np.array([
    [1, 2, 3, 0, 0],
    [2, 1, 2, 3, 0],
    [3, 2, 1, 2, 3],
    [0, 3, 2, 1, 2],
    [0, 0, 3, 2, 1]
], float)

b = np.array([14, 22, 33, 26, 22], float)

def jacobi(A, b, tol=1e-6, maxit=100000):
    n = len(b)
    x = np.zeros(n)
    for it in range(1, maxit + 1):
        x_new = x.copy()
        for i in range(n):
            s = A[i] @ x - A[i, i] * x[i]
            x_new[i] = (b[i] - s) / A[i, i]
        if np.linalg.norm(x_new - x, np.inf) < tol:
            return x_new, it
        x = x_new
    return x, maxit

def gauss_seidel(A, b, tol=1e-6, maxit=100000):
    n = len(b)
    x = np.zeros(n)
    for it in range(1, maxit + 1):
        x_old = x.copy()
        for i in range(n):
            s1 = A[i, :i] @ x[:i]
            s2 = A[i, i+1:] @ x_old[i+1:]
            x[i] = (b[i] - s1 - s2) / A[i, i]
        if np.linalg.norm(x - x_old, np.inf) < tol:
            return x, it
    return x, maxit

# Run and time
t0 = perf_counter()
xj, ij = jacobi(A, b)
t1 = perf_counter()

t2 = perf_counter()
xg, ig = gauss_seidel(A, b)
t3 = perf_counter()

print("Jacobi:       iters =", ij, "time =", t1 - t0)
print("Gauss-Seidel: iters =", ig, "time =", t3 - t2)
```
