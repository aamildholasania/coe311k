```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
```

```python
# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack design matrix with column of ones for intercept
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")
```

```python
x_line = np.linspace(0, max(X), 100)
lstsq = m * x_line + c

plt.title("Least Square Regression Between the given Data Points")
#data points
plt.scatter(X, Y)
#this is regression line
plt.plot(x_line, lstsq)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Randomize vectors X and Y
Xrand = np.random.rand(X.size)
Yrand = np.random.rand(Y.size)

idx = np.argsort(Xrand)
Xrand = Xrand[idx]
Yrand = Yrand[idx]

print("Randomized X Vector:", Xrand)
print("Randomized Y Vector:", Yrand)

B = np.vstack([Xrand, np.ones(len(Xrand))]).T
n, d = np.linalg.lstsq(B, Yrand, rcond=None)[0]

print("Slope:", n, "Intercept:", d)
```

```python
X = np.linspace(Xrand.min(), Xrand.max(), 100)
y = n * X + d

plt.title("Least Square Regression Model of Randomized Vectors")
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(Xrand, Yrand)
plt.plot(X, y)
plt.show()
```

```python
# Randomization
X1 = np.arange(0, 1.01, 0.01)
m = np.random.rand()
b = np.random.rand()

Y = m * X1 + b + np.random.rand(X1.size)

print("Random Slope:", m, "Random Intercept:", b)
```

```python
# Linear algebra
B = np.vstack([X1, np.ones(len(X1))]).T
n, d = np.linalg.lstsq(B, Y, rcond=None)[0]

print("Calculated Slope:", n, "Calculated Intercept:", d)
```

```python
#residuals
residuals = Y - (n * X1 + d)

max_residual = np.argmax(residuals)
min_residual = np.argmin(residuals)

print("Max Residual:", residuals[max_residual], "at point X =", X1[max_residual])
print("Min Residual:", residuals[min_residual], "at point X =", X1[min_residual], "\n")

#all residuals
print("Every Residual:")
for i in range(len(residuals)):
    print(residuals[i])
```

```python
plt.title("Least Square Regression Model of Own Linear Equation")
plt.xlabel("X")
plt.ylabel("Y")

plt.plot(X1, n * X1 + d)
plt.scatter(X1, Y)

plt.show()
```
